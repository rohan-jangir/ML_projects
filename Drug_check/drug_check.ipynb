# Explore the behavior of the data by using diffrent type og plot

### 1. Histograms
* A histogram is a graphical representation of the distribution of a dataset. It is used to visualize the frequency or count of data points that fall within specified ranges (bins) of a continuous variable. 

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### 2. pie chart
* A pie chart is a circular statistical graphic that is divided into slices to illustrate numerical proportions. 

```python
plt.figure(figsize=(8, 8))
df['Drug'].value_counts().plot(kind='pie',autopct='%2f')
plt.title('Distribution of Drugs')
plt.ylabel('')
plt.show()
```

### 3. Bar Plots
* A bar graph (or bar chart) is a graphical representation of data where individual bars represent the quantity or frequency of different categories

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='Sex')
plt.title('Count of Patients by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='BP')
plt.title('Count of Patients by Blood Pressure')
plt.xlabel('Blood Pressure')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(7, 4))
sns.countplot(data=df, x='Cholesterol')
plt.title('Count of Patients by Cholesterol Level')
plt.xlabel('Cholesterol Level')
plt.ylabel('Count')
plt.show()
```

### 4. Box Plots
* A box plot, also known as a whisker plot, is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.


```python
plt.figure(figsize=(7, 4))
sns.boxplot(data=df, x='Sex', y='Age')
plt.title('Distribution of Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

### 5. Heatmap

## LabelEncoder.
* The LabelEncoder in scikit-learn is a utility class used for encoding categorical features into numerical values

```python
from sklearn.preprocessing import LabelEncoder
for i in df.columns:
    if i=='Na_to_K':
        continue
    elif i=='Age':
        continue
    else:  
        a = LabelEncoder()
        new_col = a.fit_transform(df[i])
        df[i] = new_col
```

### 6. pairplot
* A pair plot, also known as a scatterplot matrix, is a useful visualization for understanding the pairwise relationships between multiple variables in a dataset. It creates a grid of scatterplots, where each plot shows the relationship between a pair of variables.

```python
sns.pairplot(df)
```

# Train the model with loaded dataset drug200 

```python
# train the model using decision tree classifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
model = DecisionTreeClassifier()
```

```python
# split data into dependent and independent data

x = df[['Age','BP','Cholesterol','Sex','Na_to_K']]
y = df['Drug']
Xtrain,Xtest,ytrain,ytest = train_test_split(x,y)
```

```python
# fitt the train and test value in decision tree classifier model

model.fit(Xtrain,ytrain)
```

```python
# predict the test values

model.predict(Xtest)
```

# Convert Drug numeric value into string value for getting output in drug name

```python
col1 = int(input('Enter First Value'))
col2 = int(input('Enter second Value'))
col3 = int(input('Enter third Value'))
col4 = int(input('Enter fourth Value'))
col5 = int(input('Enter fifth Value'))
d = pd.DataFrame([{'Age':col1,'BP':col2,'Cholesterol':col3,'Sex':col4,'Na_to_K':col5}])

for predict in model.predict(d):
    if predict==4:
        predict='DrugY'
    elif predict == 3:
        predict='DrugX'
    elif predict == 2:
        predict='DrugC'
    elif predict == 1:
        predict='DrugB'
    else:
        predict='DrugA'
    print('Your Prefered Drug is:  ',predict)
```

# Check our model score which is good or bad

```python
# check the model accuracy.

model.score(Xtrain,ytrain)
```
